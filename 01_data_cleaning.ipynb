```python
import pandas as pd

df = pd.read_csv(r"C:/Users/Administrator/Documents/business-analytics-platform/data/Details.csv")
df.head()
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
#Remove Duplicates
df = df.drop_duplicates()
```

```python
#Fix Column Names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

df.columns

df.to_csv("C:/Users/Administrator/Documents/business-analytics-platform/data/cleaned_details.csv", index=False)
```

```python
#Fix Data Types
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
```

```python
#Handle Missing Values
df = df.dropna()
```

```python
#Save Cleaned Data
df.to_csv("C:/Users/Administrator/Documents/business-analytics-platform/data/cleaned_details.csv", index=False)
```
