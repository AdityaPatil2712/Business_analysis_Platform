```python
kpis = {
    "Total Sales": df['Amount'].sum(),
    "Total Profit": df['Profit'].sum(),
    "Profit Margin (%)": (df['Profit'].sum() / df['Amount'].sum()) * 100,
    "Average Order Value": df.groupby('Order ID')['Amount'].sum().mean(),
    "Total Quantity Sold": df['Quantity'].sum()
}

kpis
```
