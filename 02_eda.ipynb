```python
import pandas as pd

df = pd.read_csv(r"C:/Users/Administrator/Documents/business-analytics-platform/data/Details.csv")
df.head()
```

```python
#Basic metrics
total_sales = df['Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()

total_sales, total_profit, total_orders
```

```python
#Sales by Category
import seaborn as sns
import matplotlib.pyplot as plt

category_sales = df.groupby('Category')['Amount'].sum().sort_values()

category_sales.plot(kind='barh', title="Sales by Category")
plt.show()
```

```python
#Profit by Sub-Category
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()

sub_profit.plot(kind='barh', figsize=(8,6))
plt.title("Profit by Sub-Category")
plt.show()
```

```python
#Payment Mode Analysis
df.groupby('PaymentMode')['Amount'].sum().plot(
    kind='pie', autopct='%1.1f%%', title='Payment Mode Share'
)
plt.ylabel("")
plt.show()
```
